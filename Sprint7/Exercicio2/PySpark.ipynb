```python
import pyspark
from pyspark.sql import SparkSession
from pyspark.sql.functions import split
import re
spark = SparkSession.builder.master("local[*]") \
.appName('WordCounter') \
.getOrCreate()
sc = spark.sparkContext
```

```python
def rev(x):
  characters='#-*`.[]():/${}1234567890"?=<>,;_~'
  for ch in characters:
    x = x.replace(ch, ' ')
  return x
```

```python
df1 = sc.textFile("README.MD")
```

```python
val2 = df1.map(lambda x: x.lower()) \
.map(rev) \
.flatMap(lambda x: x.split(" "))\
.filter(lambda x: x != "") \
.map(lambda x: (x, 1)) \
.reduceByKey(lambda x, y: x + y) \
.collect()
```

```python
for i, j in val2:
    print(f"{i}: {j}")
```
